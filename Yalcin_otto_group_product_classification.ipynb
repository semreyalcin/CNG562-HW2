Otto Group Product Classifiacation

```
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
train_url = 'https://raw.githubusercontent.com/semreyalcin/CNG562-HW2/master/Otto%20Group%20Product%20Classification/train.csv'
data = pd.read_csv(train_url)
data.head()
```

```
y = data['target']
X = data.drop(['target','id'],axis = 1)
data.head(),X.head(),y.head()
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(solver='lbfgs', 
                    alpha=1e-5,
                    hidden_layer_sizes=(25, 9), 
                    random_state=1)

mlp.fit(X_train, y_train)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,mlp.predict(X_test))
```
