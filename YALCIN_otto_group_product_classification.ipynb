Otto Group Product Classifiacation

#Preparing Data

```
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
train_url = 'https://raw.githubusercontent.com/semreyalcin/CNG562-HW2/master/Otto%20Group%20Product%20Classification/train.csv'
data = pd.read_csv(train_url)
data.head()
```

```
y = data['target']
X = data.drop(['target','id'],axis = 1)
data.head(),X.head(),y.head()
```

#Multi-Layered Perceptron

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```
from sklearn.neural_network import MLPClassifier
MLP = MLPClassifier(solver='adam', 
                    alpha=1e-5,
                    hidden_layer_sizes=(25, 15), 
                    random_state=21,
                    shuffle = True,
                    max_iter = 300,
                    early_stopping = True,
                    verbose = True)

MLP.fit(X_train, y_train)
```

```
from sklearn.metrics import accuracy_score
acc_MLP = accuracy_score(y_test, MLP.predict(X_test))
acc_MLP
```

#KNN

```
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=5).fit(X_train,y_train)
acc_KNN = accuracy_score(y_test, KNN.predict(X_test))
acc_KNN
```

# Support Vector Machine

```
from sklearn.svm import SVC 
SVM = SVC().fit(X_train,y_train)
acc_SVM = accuracy_score(y_test, SVM.predict(X_test))
acc_SVM
```

#XGBoost

```
from xgboost import XGBClassifier

XGB = XGBClassifier().fit(X_train,y_train)
acc_XGB = accuracy_score(y_test, XGB.predict(X_test))
acc_XGB
```

#Result Comparison

```
"Classifier  |  Result",
"MLP:        |  " + str(acc_MLP),
"KNN:        |  " + str(acc_KNN),
"SVM:        |  " + str(acc_SVM),
"XGB:        |  " + str(acc_XGB)
```
